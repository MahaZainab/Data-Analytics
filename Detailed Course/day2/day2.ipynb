# Tranformers
Example: Language Translation
Let's consider an example of translating an English sentence to French using a Transformer model.

Input Sentence: "I love programming."
Step 1: Tokenization and Embedding

The input sentence is tokenized into words and converted into a sequence of embeddings. Each word is represented as a vector.
Step 2: Positional Encoding

Positional encodings are added to the word embeddings to indicate their positions in the sentence.
Step 3: Encoder

The sequence of embeddings (with positional encodings) is passed through the encoder, which consists of multiple layers. Each layer applies self-attention and feed-forward neural network operations, creating a context-aware representation for each word.
Step 4: Decoder

The decoder takes the context-aware representations from the encoder and generates the output sequence one word at a time. It uses self-attention to focus on relevant parts of the input and previously generated output.
Step 5: Output Sentence

The decoder generates the translated sentence in French: "J'aime programmer."
